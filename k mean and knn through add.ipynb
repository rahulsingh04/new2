```python
df=df.drop(['Ad Topic Line' , 'Country' ,'Unnamed: 8','City'],axis = 1)
df.head()
```

```python
s = pd.DataFrame(df)
s
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```python
y =s['Clicked on Ad']
y.head()
```

```python
k=2

kmeans_model = KMeans(n_clusters = k)

kmeans_model.fit(s)
```

```python
kmeans_model.labels_
```

```python
n = pd.crosstab(y , kmeans_model.labels_)

n
```

```python
from sklearn.cross_validation import train_test_split
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier(n_neighbors =5 , metric = 'minkowski' , p=2)
```

```python
clf.fit(x_train , y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn. metrics  import confusion_matrix
```

```python
c_m = confusion_matrix(y_test ,y_pred)
```

```python
from matplotlib.colors import ListedColormap
x_set,y_set=x_train , y_train
x1,x2 = np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                    np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(x1,x2,clf.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(("red","green")))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,i],c=ListedColormap(("red","green"))(i),label=j)
    
plt.title("k-nn (Training set)")
plt.xlabel("Age")
plt.ylabel("Survived")
plt.legend()
plt.show()
```
